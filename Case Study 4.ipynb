```python
import pandas as pd 
# Input Data Excel
file_path = r'C:\Users\ASUS\Downloads\Pistachio_28_Features_Dataset.xlsx'
pista = pd.read_excel(file_path)
pista.head(3)
```

```python
# Checking Missing Value 
missing_values = pista.isnull().sum()
print("Missing Values:", missing_values)
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
pista.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# Preprocessing Data
pista= pista.drop_duplicates()
```

```python
# Handling Outliers
from scipy import stats
import numpy as np
z_scores = stats.zscore(pista.select_dtypes(include=[np.number]))
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
pista = pista[filtered_entries]
```
